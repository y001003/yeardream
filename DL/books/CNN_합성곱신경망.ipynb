### 1.1.1 합성곱 연산

합성곱신경망(CNN)은 말 그대로 **합성곱**이라는 연산을 이용하는 인공신경망을 말한다. 그렇다면 합성곱은 무엇인가라는 질문으로 연결된다. **합성곱 연산**(Convolution operation)은 대상이 되는 이미지를 **필터(filter)** 또는 **커널(kernel)** 이라는 행렬로 겹치며 훓어나간다. 그리고 각 이미지와 필터의 원소의 값을 곱한 후 모두 합한 값을 출력하는 연산이다.

### 1.1.2 패딩

Feature Map은 Convoultion layer를 거치며 크기가 작아지게 된다. 이를 막기위해 입력 데이터 외각에 지정된 픽셀만큼 특정 값으로 채워 넣는 것을 **패딩(Padding)**이라고 한다.

image.png

### 1.1.3 출력 크기 계산식

- 입력 데이터 높이 : H
- 입력 데이터 폭 : W
- 필터 높이 : FH
- 필터 폭 : FW
- Stride 크기 : S
- 패딩 사이즈 : P

$$
OutputHeight = OH = \frac{(H + 2P - FH)}{S} + 1
$$
$$
OutputWeight = OW = \frac{(W + 2P - FW)}{S} + 1
$$

## 1.2 풀링층(Pooling layer)

### 1.2.1 풀링층

image.png

풀링 층은 계산량과 메모리 사용량, 파라미터의 수를 줄이기 위해 축소본(subsample)을 만드는 것이다. 풀링 층에서는 **풀링 커널(pooling kernel)**을 사용해, 특정값만 다음층으로 넘긴다. 가장 널리쓰이는 **최대 풀링층(max pooling layer)**는 커널 내에 가장 큰 입력값(max pooling layer)만 다음 층으로 전달된다. 위의 그림은 핸즈온 머신러닝 책에 나온 것으로 max pooling layer를 보여주는 예이다. 이외에 평균 풀링 층(average pooling layer)가 있다.



### 1.2.2 풀링층 출력 크기 계산

$$
OutputRowSize = \frac{InputRowSize}{PoolingSize}
$$
$$
OutputColumnSize = \frac{InputColumnSize}{PoolingSize}
$$
