## 집합과 논리

### 집합

집합 set: 여러 원소들의 모임으로 중복된 원소를 가지지 않는 모임, 순서가 없다  

표기법
1. 원소나열볍 A = {1,2,3,4,5}
2. 조건제시법 A = {x|0<x<=5, x는 자연수}

합집합 : A의 원소들과 B의 원소들을 모두 모은 집합  
교집합 : A와 B에 동시에 속하는 원소 집합
서로소 : A와 B에 공통으로 속한 원소가 하나도 없을 경우
차집합 : A에 속하는 원소 중에서 B에 속하지 않은 원소만으로 이루어진 집합 A - A와 B 교집합
여집합 : Ac A에 속하지 않지만 전체집합에 속하는 원소 집합

집합의 명제  
: 객관적으로 참, 거짓을 판단할 수 있는 문장이나 수식.



### 집합과 논리

부정 :~p 명제 p에 대하여 p의 진릿값을 반대롤 갖는 명제  
논리곱 (and) : 명제 p,q에 대하여 p와 q가 모두 참일 경우에만 참이고, 그렇지 않을 경우 거짓  
논리합 (or) : 명제 p,q에 대하여 p와 q가 모두 거짓일 경우에만 거짓  

역 : 조건 명제 p -> q에 대하여 가정과 결론이 바뀐 q -> p
이 : 조건 명제에 대하여 가정과 결로을 각각 부정 ~p -> ~q
대우 : 가정과 결론이 바뀐 동시에 부정 ~q -> ~p

조건 명제의 대우, 역과 이는 합치한다. True or False가 같다.  
  명제 함수 P(x) :
변수 x를 포함하여 진리값을 판별할 수 있는 문장이나 수식.

## 함수

집합 A,B에 대해 모든 집합 A의 원소에 대해서 집합 B의 원소가 하나씩 대응될 때,  
f:A -> B 라고 쓰고 A에서 B로 가는 함수 f라고 읽는다.  
A의 원소 a에 대해 a에 대응되는 b의 값을 f(a)라고 쓴다. 

f:A -> B  
정의역 : f가 정의된 A를 f의 정의역이라고 한다.  
공역 : B를 f의 공역이라고 한다.  
치역 : 공역 중 f(A)로 대응되는 것들의 모임  


일대일 함수 : 정의역의 모든 원소들이 서로 다른 공역의 원소와 대응하는 함수
전사 함수 : 공역과 치역이 일치하는 함수
일대일 대응 전사함수 : 일대일 함수 중에서 공역과 일치하는 함수
항등 함수 f:A->A 인 함수  
역함수 : f:A->B 일때 g:B->A 이고 g f = f g가 항등함수가 되는 g 함수. 이를 g = f-1라고 쓴다.
상수 함수 f:A->B 일때 f의 치역이 단 하나만 존재
