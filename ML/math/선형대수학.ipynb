## 2. 벡터란 무엇인가?

#### 벡터의 정의 Definition

- 벡터 공간의 원소
- 여러 개의 **숫자** 모음
- 행 벡터 $\begin{pmatrix}1&2&3 \end{pmatrix}$
- 열 벡터 $\begin{pmatrix}1\\2\\3 \end{pmatrix}$


#### 벡터 용어
- 기저(basis) = 축(axis) 
    - x축, y축, z축
- 차원(dimension) = 벡터공간의 기저 갯수
- 원소(delement) = 벡터
- 벡터 공간 : 각 축을 기준으로 원소(벡터)들이 표시되는 공간

#### 백터 공간
- 벡터 공간(Vector Space) = 집합
- 크기(norm) = 벡터의 길이
- 방향(direction) = 단위 벡터(unit vector)

### 벡터 간 거리
데이터 사이의 **유사성** 측정 ➡ 벡터 사이의 유사성 측정 ➡ 거리 함수(distance function) 사용

- L1 distance : Manhattan Distance
    - 축을 따라서 거리 측정 ➡ 축을 중요시할 때 사용
- L2 distance : Euclidean Distance
    - 축을 무시하고 공간에서 직선 거리 ➡ 거리가 더 중요할 때
- 코사인 유사도(cosine similarity) : 
    - 각도를 비교, 각 벡터가 바라보는 방향이 얼마나 비슷한가

- L1 norm = 벡터와 원점 사이의 L1 distance
