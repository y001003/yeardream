# 머신러닝 (Machine Learning)

### Formal Definition
A computer program is said to learn from experience E with respect to some class of tasks T and performance measure P if its performance at tasks in T, as measured by P, improves with experience E.
    
          f
    T : X ➡ Y

- Task : 작업
- X : row data
- Y : result data
- f : experience learning function
- 머신러닝은 P(Performance)가 높은 f 를 찾는 것
- 직관적인 의미
    - 사람이 학습하듯 예시들(데이터)를 통해서 패턴을 찾아내는 방법
    - 컴퓨터에게 예시(데이터)와 데이터를 바라보는 방법(모델)을 주고 판단을 내리는 기준을 찾아내는 것
    - 그러나 컴퓨터는 (0,1) 밖에 이해하지 못해서, 수학을 이용해서 접근한다.
    - 이를 통해 **"prediction for unseen data"**의 performance를 높히는것

## 머신러닝의 종류
1. 지도학습(Supervised Learning)
    - 학습시 사용하는 데이터(X)가 예측대상(y:label)을 포함하고 있고, (X,y)의 pair를 학습에 사용하는 방법
    - 분류(Classification), 회귀(Regression)
2. 비지도학습(Unsupervised Learning)
    - 학습 시 사용하는 데이터만을 이용해서 학습하는 방법
    - 평가가 쉽지 않다.
    - 분배(Clustering), 차원축소(Dimensionality Reduction)
3. 강화학습(Reinforcement Learning)
    - 어떤 환경 안에서 정의된 에이전트(학습 대상)가 현재의 상태를 인식하여, 선택 가능한 행동들 중 보상을 최대화하는 행동 혹은 행동 순서를 선택하는 방법이다.
    - 가장 사람의 학습방법과 닮아있다. 중요한 것은 학습대상과 환경 변화에 대한 상호작용(interaction)이다.
    - 자율주행 자동차가 도심속에서 달릴때 주변의 사물을 인식하며, 달릴지 멈출지, 어느방향으로 가야하는지 등을 결정하는 알고리즘을 만들떄 사용된다. 
    - 강화학습 내에서도 지도학습과 비지도학습이 나뉘어진다.

### Loss Function

Loss : 예측한 결과값(Perdicted value)과 실제 정답(Target Value) 사이의 차이.  
Loss = Target value - Predicted value

ex) $L = wX + b$

$X = data$  
$w, b = parameter(=weight)$   

Loss 값을 최소한으로 줄여 0에 가까워지게 만들어야 한다.  
그런데 data는 바꿀수 없음으로 w와 b를 바꾸어 Loss 값을 조정한다.

즉 L 은 w에 대한 함수이다.

이 L가 0 또는 최소가 되는 지점의 weight를 찾는 것이 Loss Function Optimization이며 머신러닝의 학습방법이다.
