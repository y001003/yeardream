### OOP : Object-Oriented Programming 객체 지향 프로그래밍.객체 기반으로 프로그램을 만드는 방법론



- OOP는 실제 세상을 표현하고 있는 여러 개념들을 프로그램으로 옮겨 구현하기 위한 프로그래밍 패러다임 -> 개념의 추상화
- 객체(Object): 실제 사물 하나하나를 의미할 수 있고, 이러한 사물들이 공유하는 속성을 정의한 것을 클래스(Class)라고 한다.
- 추상적인 개념 Class를 실제로 사용하려면 하나하나의 개별 사물로 만들어야 하는데, 이를 개별 사물인 객체(Object)라고 한다. -> 개념의 구체화(instantiation)


OOP 장점
1. 라이브러리 형태로 재사용 쉽다. ▶ 생산성 향상
2. 자연적인 모델링
3. 상속의 개념 때문에, 재사용성이 극대화 된다.
4. 유지보수 용이성 증가

#### 상속

image.png

- 클래스는 개념의 추상화이기 때문에, 해당 개념을 계승하는 하위 개념을 만들 수 있다.
- 상속하는 클래스는 superclass, 상속받는 클래스는 subclass

#### 다형성(Polymorphism)

- 여러 하위 클래스가 같은 class method를 상속받더라고, 그 내용을 다르게 저장 가능하다.
- 예를 들어 Animal Class에 Speak()라는 메서드가 있다면, 이 기능은 다른 동물을 표현하는 subclass들마다 다르게 구현 가능하다.  ex) Dog : "Woof", Cat : "Meow", Cow : "Moo"
- 이렇게 개념이 다양항 한 형태로 표한가능한 것을 **Polymorphism**이라고 하며, 이런 다형성을 구현할 수 있는 기능을 **Method Overriding**이라고 한다.

#### 은닉화(Encapsulation)

- 사용자는 Class의 내부 구현 코드를 보지 않아도 내부 데이터와 기능을 사용하는데 문제가 없다.
- 이를 통해 잘 만들어진 open-source library들을 손쉽게 이용할 수 있다.
