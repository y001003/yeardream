```python
from PIL import Image
img = Image.open("/content/drive/MyDrive/이어드림/image/0.png")
img = img.convert('L')
img
```

- L : 흑백
- RGB : Red, Green, Blue 삼원색을 이용한 색 표현

`PngImageFile` 객체를 `ndarray`로 변환

```python
import numpy as np

arr = np.array(img)
print(type(arr))
print(arr.shape)
```

ndarray는 matplot 라이브러리로 시각화할 수 있습니다. 

```python
from matplotlib import pyplot as plt
plt.imshow(arr, cmap='gray')
```

Q. 이미지 Crop하기 (200, 200) 크기로 영상 자르기

```python
# 코드를 작성하세요

plt.imshow(arr[100:300,100:300], cmap='gray')
```

```python
arr.shape[0]
```

200보다 큰 값을 255로 변환하라. 

```python
for i in range(arr.shape[0]):
    for j in range(arr.shape[1]):
        if arr[i][j] > 200:
            arr[i][j] = 255
```

```python
plt.imshow(arr, cmap='gray')
```

동일한 영상을 가로 방향으로 이어 붙여 출력하라.

```python
arrsum = np.concatenate([arr,arr],axis=1)
```

```python
plt.imshow(arrsum, cmap='gray')
```

영상 회전 (90/180/270/360)  
`np.rot90`

```python
arr90 = np.rot90(arr)
plt.imshow(arr90, cmap='gray')
```

영상의 반전

```python
arr180 = np.rot90(arr90)
plt.imshow(arr180, cmap='gray')
```
